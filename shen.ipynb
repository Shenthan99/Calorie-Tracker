```python
food_items = {
    'bread_jam': 130,
    'cereal': 240,
    'turkey_sandwich': 300,
    'grilled_chicken_breast': 320,
    'pizza': 266,
    'spaghetti': 266,
    'fried_rice': 180,
    'salad': 120,
    'fruit': 50,
    'yogurt': 100,
    'milk': 120,
    'water': 0
}
```

```python
import datetime

# Get today's date
today = datetime.date.today()

# Get the current day of the week
weekday = today.strftime("%A")

# Define a dictionary to store daily food intake
weekly_intake = {
    "Monday": {"breakfast": None, "lunch": None, "dinner": None},
    "Tuesday": {"breakfast": None, "lunch": None, "dinner": None},
    "Wednesday": {"breakfast": None, "lunch": None, "dinner": None},
    "Thursday": {"breakfast": None, "lunch": None, "dinner": None},
    "Friday": {"breakfast": None, "lunch": None, "dinner": None},
    "Saturday": {"breakfast": None, "lunch": None, "dinner": None},
    "Sunday": {"breakfast": None, "lunch": None, "dinner": None},
}

# Ask for daily food intake
for day in weekly_intake.keys():
    print(f"Day-{list(weekly_intake.keys()).index(day) + 1}; {day}")
    weekly_intake[day]["breakfast"] = input("What's for breakfast? ")
    weekly_intake[day]["lunch"] = input("What's for lunch? ")
    weekly_intake[day]["dinner"] = input("What's for dinner? ")
```

```python
import matplotlib.pyplot as plt

# Calculate total calorie intake for each day
daily_calories = {}
for day, meals in weekly_intake.items():
    total_calories = sum([food_items[item] for item in meals.values() if item in food_items])
    daily_calories[day] = total_calories

# Calculate mean recommended calorie intake
# Replace this with actual calculation based on user parameters
mean_recommended_calories = 2000

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(daily_calories.keys(), daily_calories.values(), color='skyblue', label='Calorie Intake')
plt.axhline(y=mean_recommended_calories, color='r', linestyle='-', label='Recommended Intake')
plt.xlabel('Day of the Week')
plt.ylabel('Calories')
plt.title('Weekly Calorie Consumption')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
